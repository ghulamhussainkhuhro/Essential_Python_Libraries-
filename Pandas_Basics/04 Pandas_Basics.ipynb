# Insert function

```python
import pandas as pd

var = pd.DataFrame({"A":[1,2,3,4,5],"B":[9,8,7,6,5]})

var
```

```python
# Inserting data between a and b
var.insert(1,"New Data", var["A"])

var
```

```python
var.insert(2,"New Data 1",['a','b','c','d','e'])
var
```

```python
# We can aslo copy limited amount of data
var = pd.DataFrame({"A":[1,2,3,4,5],"B":[9,8,7,6,5]})
var["sliced data"] = var["A"][:3] # Will copy data from colum A upto 2 index and store in new Column named sliced data
var
```
