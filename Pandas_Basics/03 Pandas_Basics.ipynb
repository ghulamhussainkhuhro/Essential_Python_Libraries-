# Arithematic Operations

```python
import pandas as pd
var = pd.DataFrame({"A":[1,2,3,4], "B":[5,0,7,8]})
print(var)
```

```python
var["Sum"] = var["A"]+var["B"]
var["Difference"] = var["A"]-var["B"]
var["Product"] = var["A"]*var["B"]
var["Qoutient"] = var["A"]/var["B"]
var["A power B"] = var["A"]**var["B"]
var["A mod B"] = var["A"]%var["B"]

var
```

```python
var["Is Qouteient > 0"] = var["Qoutient"] > 0 
var
```
